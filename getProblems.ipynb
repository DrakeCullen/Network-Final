# Grab all problems and their difficulty

```python
problem_to_dict = {}
page = 0


while True:
    result = requests.get(f"https://open.kattis.com/problems?page={page}")
    soup = bs4.BeautifulSoup(result.text, "html.parser")
    
    buttons = soup.find('div', attrs={'class':"dataTables_paginate"}).findAll('a')
    
    table = soup.find('table', attrs={'class':"problem_list table sortable table-responsive table-kattis center table-hover table-multiple-head-rows table-compact"})
    table_body = table.find('tbody')
    rows = table_body.find_all('tr')

    for row in rows:
        name, diff = [row.findAll('td')[0].text, row.findAll('td')[8].text]
        problem_to_dict[name] = diff
    page += 1
    
    if buttons[1]['class'][0] == "disabled": 
        break
```
